```python
def get_scores_sc(model):
    
    # instantiate model
    model = model()
    
    # fit to scaled data
    model.fit(X_train_sc, y_train)
    
    # make predictions
    preds = model.predict(X_test_sc)

    # print results
    print(f'{model} \nTraining Score: {model.score(X_train_sc, y_train)}')
    print(f'{model} \nTesting Score: {model.score(X_test_sc, y_test)}')
    print()
    print(classification_report(y_test, preds))
    print(confusion_matrix(y_test, preds))

    return
```

```python
#importing required packages
import keras
from keras.layers import Input, Dense
from keras import regularizers
import tensorflow as tf
from keras.models import Model
from sklearn import metrics
from tensorflow.keras import layers, losses
import numpy as np
import pandas as pd
import tensorflow as tf
import matplotlib.pyplot as plt
from sklearn.metrics import accuracy_score
from tensorflow.keras.optimizers import Adam
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras import Model, Sequential
from tensorflow.keras.layers import Dense, Dropout
from sklearn.model_selection import train_test_split
from tensorflow.keras.losses import MeanSquaredLogarithmicError
```

```python
from tensorflow.keras.layers import Dense, Dropout

class AutoEncoder(Model):
 

  def __init__(self, output_units):
    super().__init__()
    self.encoder = Sequential([
      Dense(output_units, activation='relu'),
      Dense(64, activation='relu'),
      Dense(output_units, activation='relu'),
      Dense(32, activation='relu'),
      
    ])
    self.decoder = Sequential([
      Dense(16, activation='relu'),   
      Dense(output_units, activation='sigmoid'),
      Dense(output_units, activation='relu'),
      Dense(70, activation='relu'),
    ])
  
  def call(self, inputs):
    encoded = self.encoder(inputs)
    decoded = self.decoder(encoded)
    return decoded
  
modeland = AutoEncoder(output_units=X_train_sc.shape[1])
# configurations of model
modeland.compile(optimizer='adam', loss='msle',  metrics=['accuracy'])
```
